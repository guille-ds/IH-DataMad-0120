```python
import json
import requests
import os
from dotenv import load_dotenv
```

```python
load_dotenv()
token=os.getenv("githtok")
```

```python
def requestGithub(endpoint):
    token = os.getenv("githtok")
    if not token:
        raise ValueError("You must set a GITHUB_APIKEY token")
    
    baseUrl = "https://api.github.com"
    url = baseUrl+endpoint
    print(f"Pidiendo movidas a  {url}")
    headers = {
        "Authorization": f"token {token}"
    }
    res = requests.get(url,headers=headers)
    if res.status_code != 200:
        print(res.text)
        raise ValueError("Bad Response")
    return res.json()
```

```python
contents = requestGithub("/repos/ironhack-datalabs/scavenger/contents?extension:scavengerhunts")
print("Hay ", len(contents) ," archivos con la extensión .scavengerhunts")
```

```python
names=[]
for e in contents:
    n = e["name"]
    names.append(n)
print(names)
```

```python
carpetas = names[1:]
print(carpetas)
```

```python
def request(x):
    token = os.getenv("githtok")
    if not token:
        raise ValueError("You must set a GITHUB_APIKEY token")
    
    headers = {
        "Authorization": f"token {token}"
    }
    
    url = f'https://api.github.com/repos/ironhack-datalabs/scavenger/contents/{x}'
    res = requests.get(url,headers=headers)
    print(f"Requesting data from {url}")
    
    if res.status_code != 200:
        print(res.text)
        raise ValueError("Bad Response")
    return res.json()
    
```

```python
contenido = list(map(request,carpetas))
```

```python
contenido
```

```python
listapath = [ar["path"] for h in contenido for ar in h]
```

```python
hola = [a for a in listapath if ".scavengerhunt" in a]
```

```python
print(hola)
```

```python
ordenado = sorted(hola,key=lambda x: (x[9],x[10]))
```

```python
archivos = list(map(request,ordenado))
```

```python
print(len(archivos))
```

```python
contenido = [algo["content"] for algo in archivos]
```

```python
import base64
decod = [base64.b64decode(b) for b in contenido]
print(decod)
```

```python
final = [i.decode("UTF-8") for i in decod]
print(final)
```

```python
texto = [z.replace("\n"," ")for z in final]
```

```python
FIN = "".join(texto)
print(FIN)
```
