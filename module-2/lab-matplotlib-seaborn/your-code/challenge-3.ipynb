```python
# your code here
lista = [10,20,50]
colores = ["red", "green", "blue"]
fig,ax = plt.subplots(figsize = (15,5), nrows=1, ncols=len(lista))
for i, e  in enumerate(lista):
    plt.subplot(1,len(lista),i+1)
    plt.hist(data, bins=e, facecolor=colores[i], alpha=0.5);
```
